```python
# Q1: Create a simple line plot of y = 2x + 3 for x values ranging from 0 to 10.
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100)
y = 2*x+3

y = np.array(y)

plt.plot(x,y)

plt.title("simple line plot of y = 2x+3 ")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
# Q2: Plot sin(x) and cos(x) on the same graph with different colors, line styles, and labels, and display a legend.
x = np.linspace(0,10,50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x , y1, color='red',linestyle='--',marker='o',label='sin(x)')
plt.plot(x , y2, color='blue',linestyle='-',marker='x',label='cos(x)')

plt.title("customized line plot")
plt.xlabel("x value")
plt.ylabel("y value")

plt.grid(True)
plt.legend()

plt.show()
```

```python
# Q3: Create a bar chart showing the sales of four products: A, B, C, and D with random values.
products = ['A','B','C','D']
sales = np.random.randint(10, 100, size=4)

plt.bar(products , sales, color='green')

plt.title('bar chart example')
plt.xlabel("product")
plt.ylabel("sales")

plt.show()
```

```python
# Q4: Generate a scatter plot with random x and y values between 0 and 1, and use different colors for points.
x = np.random.rand(100)
y = np.random.rand(100)

colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.colorbar()  # add color scale
plt.show()
```

```python
# Q5: Plot a histogram for 1000 random numbers generated from a normal distribution.
data = np.random.randn(1000)

plt.hist(data, bins=30, color='green',edgecolor='black',alpha=0.7)

plt.title('Histoggram example')
plt.xlabel("data range")
plt.ylabel("Frequency")

plt.show()
```

```python
# Q6: Draw a pie chart representing the percentage of five fruits: Apple, Banana, Mango, Grapes, and Orange.
labels = ['apple','banana','mango','grapes','orange']
sizes = [30,20,25,25,56]
colors = ['red','yellow','pink','orange','purple']

plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%',startangle=90, shadow=True)

plt.title("pie chart example")

plt.show()
```

```python
# Q7: Create two subplots in one figure  one showing a sine wave and the other showing a cosine wave
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,y1,'r--')
plt.title('sin(x)')

plt.subplot(1,2,2)
plt.plot(x,y2,'b--')
plt.title('cos(x)')

plt.suptitle("multiple plot example")
plt.tight_layout()
plt.show()


plt.savefig('matplotlib-library-expontial_plot.png', dpi=300,bbox_inches='tight')

plt.show()
```

```python
# Q8: Customize a line plot by adding title, axis labels, grid lines, and saving it as 'line_plot.png'
x = np.linspace(0,10,100)
y = np.linspace(0,10,100)

y = np.array(y)

plt.plot(x,y)

plt.title("simple line plot  ")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid(True)
plt.legend()

plt.savefig('line_plot.png', dpi=300,bbox_inches='tight')


plt.show()
```

```python
# Q9: Create a 2x2 grid of subplots showing different plots: line, bar, scatter, and histogram.
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x), 'r')
axs[0, 0].set_title("Sine Wave")

# Bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
axs[0, 1].bar(categories, values)
axs[0, 1].set_title("Bar Chart")

# Histogram
data = np.random.randn(1000)
axs[1, 0].hist(data, bins=25, edgecolor='black')
axs[1, 0].set_title("Histogram")

# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
axs[1, 1].scatter(x, y)
axs[1, 1].set_title("Scatter Plot")

plt.tight_layout()
plt.show()
```

```python
# Q10: Create a line plot for y = e^x for x between 0 and 5, and change the color, marker, and line style
x = np.linspace(0,5,100)
y = np.exp(x)

plt.plot(x,y,'m-')
plt.title("exponential function")
plt.xlabel('X')
plt.ylabel('exp(x)')

plt.show()
```
