```python
# Q4: Create a new column 'TotalAmount' by multiplying Quantity and Price
# Add another column 'DiscountedPrice' that applies 10% discount on orders above $100
# Use conditional selection to filter orders where TotalAmount > $150

df["TotalAmount"] = df["Quantity"] * df["Price"]

df["DiscountedPrice"] = df["TotalAmount"].where(
    df["TotalAmount"] <= 8000,
    df["TotalAmount"] * 0.9
)

print(df[df['TotalAmount'] > 12000])


```

```python
# Q5: Select only the columns: Product, Category, TotalAmount, and Rating
# Filter and display all orders from the 'Electronics' category with Rating >= 4

selected_df = df[['Product','Category','TotalAmount','Rating']]

result = selected_df[
    (selected_df['Category'] == 'Electronics') &
    (selected_df['Rating'] >= 4)
]

print(result)
```

```python
# Q6: Use boolean indexing to find all orders where:
# - Quantity > 2 AND Price < 50 OR Rating == 5
# Display the count of such orders

filtered_orders = df[
    ((df["Quantity"] > 2) & (df["Price"] < 50)) |
    (df["Rating"] == 5)
]

count_orders = filtered_orders.shape[0]
print("Number of matching orders:", count_orders)
```

```python
# Q7: Sort the entire DataFrame by OrderDate (descending) and then by TotalAmount (descending)
# Create a new column 'SalesRank' that ranks products by TotalAmount in descending order

print(df.sort_values(by=['OrderDate', 'TotalAmount'], ascending=[False, False]))

df['SaleRank'] = df['TotalAmount'].rank(ascending=False)
print(df)


```

```python
# Q8: Find the top 5 best-selling products based on total quantity sold
# Display the product name and total quantity in descending order

product_sales = df.groupby("Product")["Quantity"].sum()

product_sales_sorted = product_sales.sort_values(ascending=False)

top_5_products = product_sales_sorted.head(5)

print(top_5_products)
```

```python
# Q9: Group the data by 'Category' and calculate:
# - Total sales (sum of TotalAmount)
# - Average rating
# - Number of orders
# Display the results sorted by total sales

category_summary = df.groupby("Category").agg(
    TotalSales=("TotalAmount", "sum"),
    AverageRating=("Rating", "mean"),
    NumberOfOrders=("OrderID", "count")
)

category_summary_sorted = category_summary.sort_values(
    by="TotalSales",
    ascending=False
)

print(category_summary_sorted)

```

```python
# Q10: Create a pivot table showing average Price for each Category and Rating combination
# Use groupby with multiple columns (Category, Product) and aggregate multiple functions

pivot_avg_price = pd.pivot_table(
    df,
    values="Price",
    index="Category",
    columns="Rating",
    aggfunc="mean"
)

print(pivot_avg_price)

category_product_summary = df.groupby(
    ["Category", "Product"]
).agg(
    TotalQuantity=("Quantity", "sum"),
    TotalSales=("TotalAmount", "sum"),
    AveragePrice=("Price", "mean"),
    AverageRating=("Rating", "mean")
)

print(category_product_summary)

```

```python
# Q12: Create a 1D NumPy array of 100 random prices between 10 and 500
# Calculate mean, median, standard deviation, min, and max using NumPy functions
# Reshape this array into a 10x10 matrix

import numpy as np

# Create a 1D NumPy array of 100 random prices between 10 and 500
prices = np.random.uniform(10, 500, 100)

# Calculate statistics
mean_price = np.mean(prices)
median_price = np.median(prices)
std_price = np.std(prices)
min_price = np.min(prices)
max_price = np.max(prices)

# Display statistics
print("Mean:", mean_price)
print("Median:", median_price)
print("Standard Deviation:", std_price)
print("Minimum:", min_price)
print("Maximum:", max_price)

# Reshape into a 10x10 matrix
price_matrix = prices.reshape(10, 10)

print("\n10x10 Price Matrix:")
print(price_matrix)

```

```python
# Q14: Create NumPy arrays using special values:
# - Array of zeros (5x5)
# - Array of ones (3x4)
# - Identity matrix (4x4)
# - Array with values from 0 to 50 with step of 5 using arange
# Concatenate and stack these arrays
zeros_array = np.zeros((5, 5))
ones_array = np.ones((3, 4))
identity_array = np.eye(4)
range_array = np.arange(0, 51, 5)

print("Zeros Array:\n", zeros_array)
print("\nOnes Array:\n", ones_array)
print("\nIdentity Matrix:\n", identity_array)
print("\nRange Array:\n", range_array)

concat_array = np.concatenate((identity_array, identity_array), axis=0)
print("\nConcatenated Identity Matrices:\n", concat_array)

vstack_array = np.vstack((identity_array, identity_array))
hstack_array = np.hstack((identity_array, identity_array))

print("\nVertical Stack:\n", vstack_array)
print("\nHorizontal Stack:\n", hstack_array)

```
